## What are the total sales and profit by Category and Sub-Category?



### Group and Summarize Data

```python
# Group by Category and Sub-Category, then sum Sales and Profit
category_summary = df.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit']].sum().reset_index()

# Sort by Sales or Profit to see top performers
category_summary = category_summary.sort_values(by='Sales', ascending=False)

category_summary.head(10)
```

## How do monthly sales and profits trend over time?

```python
# Convert 'Order Date' column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Create a new column 'Month_Year' for grouping
df['Month_Year'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
# Group by Month_Year and calculate total Sales and Profit
monthly_trend = df.groupby('Month_Year')[['Sales', 'Profit']].sum().reset_index()

# Convert Month_Year back to datetime for proper sorting in plot
monthly_trend['Month_Year'] = pd.to_datetime(monthly_trend['Month_Year'])
```

### Sales and profit over time

```python
# Plot sales and profit over time
plt.figure(figsize=(14,6))
sns.lineplot(data=monthly_trend, x='Month_Year', y='Sales', label='Sales')
sns.lineplot(data=monthly_trend, x='Month_Year', y='Profit', label='Profit')
plt.title('Monthly Sales and Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Monthly Sales and Profit Trend
From 2014 to 2017, sales have shown a clear upward trend with seasonal spikes, especially towards the end of each year—likely due to holiday shopping.
However, profits remain relatively flat and volatile, occasionally dipping into negative, which suggests:

High-cost items with low margin are being sold.

Discounts or operational inefficiencies may be affecting profitability.


## Which Sub-Categories are Most and Least Profitable?

```python
# Group by Sub-Category and sum Profit
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().reset_index()

# Sort by profit
subcat_profit_sorted = subcat_profit.sort_values(by='Profit', ascending=False)
subcat_profit_sorted.head(10)  # Top 10 profitable
```

### Profit by Sub-Category

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=subcat_profit_sorted,
    x='Profit',
    y='Sub-Category',
    palette='Spectral',
    hue='Sub-Category',
    dodge=False,
    legend=False
)
plt.title('Total Profit by Sub-Category')
plt.xlabel('Profit ($)')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

#### 📈 Top Profit-Makers:

Copiers, Phones, and Accessories are the most profitable sub-categories.

These categories likely have high margins or strong sales volume.

#### 📉 Loss-Makers:

Tables, Bookcases, and Supplies are generating negative profit.

Tables, in particular, stands out with the largest loss, which could indicate:

Overstocking or high returns

Low pricing or excessive discounts

High shipping or storage costs

#### 📌 Recommendation:

Investigate loss-making sub-categories for cost reduction, pricing adjustments, or supplier negotiations.

Focus marketing efforts on high-performing categories to boost growth.

## Analyze Profit by State/Region

```python
# Profit by Region
region_profit = df.groupby('Region')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)

# Profit by State
state_profit = df.groupby('State')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=region_profit, x='Profit', y='Region', palette='viridis')
plt.title('Total Profit by Region')
plt.xlabel('Profit ($)')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
# Top 10 profitable states
top_states = state_profit.head(10)

# Bottom 10 states (including losses)
bottom_states = state_profit.tail(10)

# Top 10 States
plt.figure(figsize=(10,5))
sns.barplot(data=top_states, x='Profit', y='State', palette='Greens_d')
plt.title('Top 10 States by Profit')
plt.xlabel('Profit ($)')
plt.ylabel('State')
plt.tight_layout()
plt.show()

# Bottom 10 States
plt.figure(figsize=(10,5))
sns.barplot(data=bottom_states, x='Profit', y='State', palette='Reds_d')
plt.title('Bottom 10 States by Profit')
plt.xlabel('Profit ($)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

🏠 Home Office segment lags behind, with the lowest profitability.

Consider marketing more aggressively to Consumers, while re-evaluating pricing, product offerings, or service strategies for Home Office customers to boost margin.
